```python
#importig libraries 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the data
df=pd.read_csv("Absenteeism_data.csv")
```

```python
df.head()
```

```python
#since 'ID' is nominal data and may effect the precision of the result it is better to drop
df=df.drop(['ID'],axis=1)
df.head()
```

```python
#Dealing with Categorical data
df['Reason for Absence'].unique()
```

```python
len(df['Reason for Absence'].unique())
```

```python
#getting dummy variables for categorical data
reasons=pd.get_dummies(df['Reason for Absence'])
reasons
```

```python
#exploring categorical values
reasons['check']=reasons.sum(axis=1)
reasons
```

```python
reasons['check'].unique()
```

```python
reasons=reasons.drop(['check'], axis=1)
```

```python
reasons=pd.get_dummies(df['Reason for Absence'],drop_first=True)
reasons
```

```python
#since adding all the columns will increase complexity grouping the reasons
#From the data description the reasons mentions are of different types
#Reasons 1 to 14 are diffrent types of diseases 
reason1=reasons.loc[:,1:14].max(axis=1) 
#15 to 17 related to pregnancy 
reason2=reasons.loc[:,15:17].max(axis=1) 
#18 to 21 signs not categoriesed 
reason3=reasons.loc[:,18:21].max(axis=1) 
#22 to 28 are light reasons
reason4=reasons.loc[:,22:28].max(axis=1) 
```

```python
#dropping 'Reasons for Absence'
df=df.drop(['Reason for Absence'],axis=1)
```

```python
df=pd.concat([df,reason1,reason2,reason3,reason4],axis=1)
df.head()
```

```python
#renaming the columns
df.columns.values
```

```python
column_names=['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'reason1', 'reason2', 'reason3', 'reason4']
df.columns=column_names
df.head()
```

```python
df=df.drop('Date',axis=1)
```

```python
df1=df.copy()
```

```python
columns_reordered=['reason1', 'reason2', 'reason3', 'reason4','Education','Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Children', 'Pets', 'Absenteeism Time in Hours',]
df1=df1[columns_reordered]
df1.head()
```

```python
df1['Absenteeism Time in Hours'].median()
```

```python
df1['Absenteeism Time in Hours']=np.where(df1['Absenteeism Time in Hours']>df1['Absenteeism Time in Hours'].median(),1,0)
```

```python
x=df1.iloc[:,0:12].values
y=df1.iloc[:,12].values
y
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size = 0.8,)
```

```python
X_train.shape
y_train.shape
```

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter=1000)
classifier.fit(X_train, y_train)
```

```python
from sklearn import metrics
classifier.score(X_train, y_train)
```
